EXTRACT TABLE 6 — Sector snapshot

```python
# Extract only the rows that belong to Table 6 within the corporation tax file
t6_raw = ct_raw[ct_raw['SourceSheet'] == 'Table_6'].copy()
# Identify which columns actually contain data for Table 6
non_null = [col for col in ct_raw.columns if t6_raw[col].notna().any()]
print("Non-null columns in Table_6:")
for c in non_null:
    print(" ", c)
```

 Identify the header row and data rows

```python
# Locate the row that contains the header labels for Table 6
# We search for the word 'Industry' which appears in the header row
header_mask = t6_raw.apply(lambda row: row.astype(str).str.contains('Industry').any(), axis=1)
print("Header row index:", t6_raw[header_mask].index.tolist())
display(t6_raw[header_mask])  # Show the header row identified by the mask
```

Building the clean Table 6 dataframe

```python
# Build the cleaned Table 6 DataFrame from the raw data
# Keep only columns that have values for the Table 6 rows
t6 = t6_raw[non_null].copy()

# Define the sector description column from the sheet
sector_col = 'Profits, losses and tax liability by Standard Industrial Classification (SIC) section, financial year 2023-24'

# Filter to rows where the sector column starts with a sector code like 'A.' or 'B.'
sector_mask = t6[sector_col].astype(str).str.match(r'^[A-U]\\.')
t6_sectors = t6[sector_mask].copy()

# Rename the generic 'Unnamed' columns to meaningful names
t6_sectors = t6_sectors.rename(columns={
    'Unnamed: 1':  'companies_filing_ct_return',
    'Unnamed: 2':  'companies_with_tax_payable',
    'Unnamed: 3':  'total_profits_gbp_m',
    'Unnamed: 4':  'trading_losses_gbp_m',
    'Unnamed: 5':  'capital_allowances_gbp_m',
    'Unnamed: 6':  'losses_claimed_gbp_m',
    sector_col:    'sector',
    'Unnamed: 7':  'group_relief_gbp_m',
    'Unnamed: 8':  'ct_payable_gbp_m'
})

# Drop the SourceSheet column since all rows are now from Table 6
t6_sectors = t6_sectors.drop(columns=['SourceSheet'])

# Convert the columns that should contain numbers into numeric type
num_cols_t6 = [
    'companies_filing_ct_return', 'companies_with_tax_payable',
    'total_profits_gbp_m', 'trading_losses_gbp_m',
    'capital_allowances_gbp_m', 'losses_claimed_gbp_m',
    'group_relief_gbp_m', 'ct_payable_gbp_m'
 ]
for col in num_cols_t6:
    t6_sectors[col] = pd.to_numeric(t6_sectors[col], errors='coerce')

# Extract a short sector code and the clean sector name
t6_sectors['sector_code'] = t6_sectors['sector'].str.extract(r'^([A-U])\\.')
t6_sectors['sector_name'] = t6_sectors['sector'].str.replace(r'^[A-U]\\.\\s*', '', regex=True)

# Reset the index after filtering rows
t6_sectors = t6_sectors.reset_index(drop=True)
display(t6_sectors[['sector_code', 'sector_name', 'companies_with_tax_payable',
                     'total_profits_gbp_m', 'ct_payable_gbp_m']])
```

Derive sector-level ratios

```python
# Deriving useful ratios for the sector snapshot table
# Calculating corporate tax payable per company in thousands of pounds
t6_sectors['ct_per_company_gbp_k'] = (
    (t6_sectors['ct_payable_gbp_m'] * 1000) / t6_sectors['companies_with_tax_payable']
).round(1)

# Calculating the effective tax rate: tax paid relative to profits
t6_sectors['effective_ct_rate_pct'] = (
    t6_sectors['ct_payable_gbp_m'] / t6_sectors['total_profits_gbp_m'] * 100
).round(2)

# Calculating the loss ratio: trading losses as a percentage of profits
t6_sectors['loss_ratio_pct'] = (
    t6_sectors['trading_losses_gbp_m'] / t6_sectors['total_profits_gbp_m'] * 100
).round(2)

display(t6_sectors[['sector_code', 'sector_name', 'ct_payable_gbp_m',
                     'ct_per_company_gbp_k', 'effective_ct_rate_pct', 'loss_ratio_pct']])
```

Tax liability time series (by SIC division)

```python
# Extracting the raw Table 8 rows from the corporation tax dataset
t8_raw = ct_raw[ct_raw['SourceSheet'] == 'Table_8'].copy()
non_null_8 = [col for col in ct_raw.columns if t8_raw[col].notna().any()]

# Inspecting the header row by searching for the label that identifies the SIC code column
header_mask_8 = t8_raw['Unnamed: 1'].astype(str).str.contains('SIC 2007 Code')
print("Header row(s):")
display(t8_raw[header_mask_8])
```

Building the time series dataframe

```python
# Builds a clean time series table from the raw Table 8 rows
# Keeps only rows whose first column looks like a SIC division code
data_mask_8 = t8_raw['Unnamed: 1'].astype(str).str.match(r'^\d')
t8_data = t8_raw[data_mask_8][non_null_8].copy()

# The human-readable column name for the sector description
year_series_col = 'Tax liability by Standard Industrial Classification (SIC) division, 2018-19 to 2023-24'

# Renames the generic columns to meaningful names
t8_data = t8_data.rename(columns={
    'Unnamed: 1':      'sic_division_code',
    'Unnamed: 2':      'sic_division_name',
    'Unnamed: 3':      'liability_2018_19',
    'Unnamed: 4':      'liability_2019_20',
    'Unnamed: 5':      'liability_2020_21',
    'Unnamed: 6':      'liability_2021_22',
    'Unnamed: 7':      'liability_2022_23',
    'Unnamed: 8':      'liability_2023_24',
    year_series_col:   'sector_section'
})

# Removes the sheet identifier column since this subset is already only Table 8 rows
t8_data = t8_data.drop(columns=['SourceSheet'])

# Converts the liability columns to numeric values and treat confidential markers as missing data
year_cols = ['liability_2018_19', 'liability_2019_20', 'liability_2020_21',
             'liability_2021_22', 'liability_2022_23', 'liability_2023_24']
for col in year_cols:
    t8_data[col] = t8_data[col].replace({'[X]': np.nan, '[x]': np.nan})
    t8_data[col] = pd.to_numeric(t8_data[col], errors='coerce')

# Extracts the broader section code from the sector description
t8_data['sector_code'] = t8_data['sector_section'].str.extract(r'^([A-U])\.')

# Resets the index after filtering and cleaning
t8_data = t8_data.reset_index(drop=True)
print("Shape:", t8_data.shape)
display(t8_data.head(10))
```

Aggregate from SIC division up to SIC section (to match Table 6)

```python
# Aggregates the SIC division liabilities up to the broader SIC section level
t8_section = (
    t8_data
    .groupby('sector_code')[year_cols]
    .sum(min_count=1)   # min_count=1 preserves NaN when all values are missing
    .reset_index()
)

# Calculates the percentage change in liability from 2018-19 to 2023-24
t8_section['liability_change_pct'] = (
    (t8_section['liability_2023_24'] - t8_section['liability_2018_19'])
    / t8_section['liability_2018_19'] * 100
).round(1)

print("Section-level time series:")
display(t8_section)
```

Company count time series

```python
# Extracting the company count time series from Table 9
# Keep only rows from the raw file that belong to Table 9
t9_raw = ct_raw[ct_raw['SourceSheet'] == 'Table_9'].copy()
non_null_9 = [col for col in ct_raw.columns if t9_raw[col].notna().any()]

# Data rows start with a SIC division code
data_mask_9 = t9_raw['Unnamed: 1'].astype(str).str.match(r'^\d')
t9_data = t9_raw[data_mask_9][non_null_9].copy()

company_count_col = 'Number of companies liable to pay tax by Standard Industrial Classification (SIC) division, 2018-19 to 2023-24'

# Rename the raw columns to readable names
t9_data = t9_data.rename(columns={
    'Unnamed: 1':        'sic_division_code',
    'Unnamed: 2':        'sic_division_name',
    'Unnamed: 3':        'companies_2018_19',
    'Unnamed: 4':        'companies_2019_20',
    'Unnamed: 5':        'companies_2020_21',
    'Unnamed: 6':        'companies_2021_22',
    'Unnamed: 7':        'companies_2022_23',
    'Unnamed: 8':        'companies_2023_24',
    company_count_col:   'sector_section'
})

# Drops the sheet identifier since we already have the relevant subset
t9_data = t9_data.drop(columns=['SourceSheet'])

# Converts the company count columns to numeric values
company_year_cols = ['companies_2018_19', 'companies_2019_20', 'companies_2020_21',
                     'companies_2021_22', 'companies_2022_23', 'companies_2023_24']
for col in company_year_cols:
    t9_data[col] = t9_data[col].replace({'[X]': np.nan, '[x]': np.nan})
    t9_data[col] = pd.to_numeric(t9_data[col], errors='coerce')

# Extracts the section code from the sector section text
t9_data['sector_code'] = t9_data['sector_section'].str.extract(r'^([A-U])\.')

# Aggregates division counts to the sector section level
t9_section = (
    t9_data
    .groupby('sector_code')[company_year_cols]
    .sum(min_count=1)
    .reset_index()
)

# Calculates percentage change in the number of companies from 2018-19 to 2023-24
t9_section['company_count_change_pct'] = (
    (t9_section['companies_2023_24'] - t9_section['companies_2018_19'])
    / t9_section['companies_2018_19'] * 100
).round(1)

display(t9_section)
```

Extract CT gap time series

```python
# Extracts the CT gap table from the tax gap dataset
t51_raw = tg_raw[tg_raw['SourceSheet'] == 'Table 5.1'].copy()
non_null_51 = [col for col in tg_raw.columns if t51_raw[col].notna().any()]

# Inspects the header row to confirm the layout of the table
header_51 = t51_raw[non_null_51].iloc[3]  # row 3 appears to be the header row
print("Header row values:")
print(header_51.values)
```

Parse the CT gap table into long format

```python
# Parsing the CT gap table into a clean DataFrame
# Keep only rows that contain actual component names and exclude the header row
t51 = t51_raw[non_null_51].copy()
component_col = 'Table 5.1: Corporation Tax gap (£ billion)'
data_51 = t51[t51[component_col].notna() & (t51[component_col] != 'Component')].copy()

# Lists the year labels for the table columns
year_labels = [
    '2005-06','2006-07','2007-08','2008-09','2009-10','2010-11','2011-12',
    '2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19',
    '2019-20','2020-21','2021-22','2022-23','2023-24'
 ]
year_cols_51 = [f'Unnamed: {i}' for i in range(1, 20)]

# Selects only the relevant columns and rename them for clarity
data_51 = data_51[['SourceSheet', component_col] + year_cols_51].copy()
data_51.columns = ['SourceSheet', 'component'] + year_labels

# Converts year values to numeric and remove any percent symbols
for yr in year_labels:
    data_51[yr] = data_51[yr].astype(str).str.replace('%', '', regex=False)
    data_51[yr] = pd.to_numeric(data_51[yr], errors='coerce')

# Removes the helper SourceSheet column before analysis
data_51 = data_51.drop(columns=['SourceSheet']).reset_index(drop=True)
print("CT gap components found:")
print(data_51['component'].tolist())
display(data_51)
```

Pivot to long format for dashboard use

```python
# Pivots the CT gap table from wide format to long format
ct_gap_long = data_51.melt(
    id_vars='component',
    var_name='year',
    value_name='value_gbp_bn'
 )

# Keeps only the main CT components we need for analysis
keep_components = ['Gross tax gap', 'Net tax gap', 'Compliance yield',
                   'Liabilities', 'Proportion of liabilities %']
ct_gap_long = ct_gap_long[ct_gap_long['component'].isin(keep_components)]

print("Shape:", ct_gap_long.shape)
display(ct_gap_long.head(20))
```

Extract customer segment gap time series (Table 1.4)

```python
# Extracts the customer segment time series from Table 1.4
t14_raw = tg_raw[tg_raw['SourceSheet'] == 'Table 1.4'].copy()
non_null_14 = [col for col in tg_raw.columns if t14_raw[col].notna().any()]

# Keeps only the columns that contain actual data for this table
t14 = t14_raw[non_null_14].copy()

segment_col = 'Table 1.4: Tax gap time series by customer group, percentage of total theoretical liabilities and £ billion'

# Filters out header rows and keeps only rows with valid segment information
data_14 = t14[t14[segment_col].notna() &
              ~t14[segment_col].isin(['Customer group', 'Tax gap as percentage of liability or £bn'])].copy()

# Renames the columns so the year columns are easy to work with
data_14.columns = ['SourceSheet', 'metric'] + year_labels + ['segment']

# Converts year values to numeric and removes percent symbols where present
for yr in year_labels:
    data_14[yr] = data_14[yr].astype(str).str.replace('%', '', regex=False)
    data_14[yr] = pd.to_numeric(data_14[yr], errors='coerce')

data_14 = data_14.drop(columns=['SourceSheet']).reset_index(drop=True)
print("Segments and metrics found:")
display(data_14[['segment', 'metric']])
```

Pivoting to long format

```python
# Reshaping the customer segment gap table so each row is one segment-year observation
segment_gap_long = data_14.melt(
    id_vars=['segment', 'metric'],
    var_name='year',
    value_name='value'
 )

# Separates percentage metrics from billion-pound metrics
seg_pct = segment_gap_long[
    segment_gap_long['metric'].str.contains('percentage', case=False)
].rename(columns={'value': 'gap_pct_of_liabilities'}).drop(columns='metric')

seg_bn = segment_gap_long[
    segment_gap_long['metric'].str.contains('£bn', case=False)
].rename(columns={'value': 'gap_gbp_bn'}).drop(columns='metric')

# Merges the two metric types into one table by segment and year
segment_gap_ts = seg_pct.merge(seg_bn, on=['segment', 'year'], how='outer')

print("Shape:", segment_gap_ts.shape)
display(segment_gap_ts.head(15))
```

Assembling three focused analytical tables

Table A — Sector Risk Snapshot (cross-sectional, 2023-24)

This answers: Which sectors generate the most CT per company, and where does HMRC concentrate its activity?

```python
# Builds Table A: a sector snapshot for the most recent year
# Starts from the cleaned sector-level data extracted from Table 6
table_a = t6_sectors[[
    'sector_code', 'sector_name',
    'companies_filing_ct_return', 'companies_with_tax_payable',
    'total_profits_gbp_m', 'ct_payable_gbp_m',
    'trading_losses_gbp_m',
    'ct_per_company_gbp_k', 'effective_ct_rate_pct', 'loss_ratio_pct'
 ]].copy()

# Adds the latest sector liability and company count metrics from the time series tables
t8_latest = t8_section[['sector_code', 'liability_2023_24', 'liability_change_pct']]
table_a = table_a.merge(t8_latest, on='sector_code', how='left')
t9_latest = t9_section[['sector_code', 'companies_2023_24', 'company_count_change_pct']]
table_a = table_a.merge(t9_latest, on='sector_code', how='left')

# Excludes the special 'U' sector because it is not a normal economic sector
table_a = table_a[table_a['sector_code'] != 'U']

# Calculates each sector's share of total corporate tax payable
total_ct = table_a['ct_payable_gbp_m'].sum()
table_a['ct_share_pct'] = (table_a['ct_payable_gbp_m'] / total_ct * 100).round(2)

# Ranks sectors by CT payable, where 1 means highest CT payable
table_a['ct_rank'] = table_a['ct_payable_gbp_m'].rank(ascending=False).astype(int)

# Sorts sectors so the largest CT contributors are at the top
table_a = table_a.sort_values('ct_payable_gbp_m', ascending=False).reset_index(drop=True)

print("Table A shape:", table_a.shape)
display(table_a)
```

Spot checking key figures

Table B — Customer Segment Dashboard Table (cross-sectional, 2024-25)

This answers: Which customer segments have the highest tax gap and yield efficiency?

Building Table B from cleaned customer data

```python
# Builds Table B: customer segment metrics for 2024-25
# Selects only the columns needed for the summary table
table_b = df_cust[[
    'segment', 'population',
    'paye_nics_receipts_gbp_bn', 'other_receipts_gbp_bn',
    'compliance_yield_gbp_bn', 'compliance_spend_gbp_m',
    'tax_gap_gbp_bn',
    'yield_pct_of_gap', 'spend_per_pound_yield', 'tax_gap_per_entity_gbp'
 ]].copy()

# Adding a total receipts column from PAYE/NICs and other receipts
table_b['total_receipts_gbp_bn'] = (
    table_b['paye_nics_receipts_gbp_bn'] + table_b['other_receipts_gbp_bn']
)
# Calculates the tax gap as a percentage of total receipts
table_b['gap_pct_of_receipts'] = (
    table_b['tax_gap_gbp_bn'] / table_b['total_receipts_gbp_bn'] * 100
).round(2)

# Defining the display order for customer segments
segment_order = {
    'Individual customers': 1,
    'Wealthy individuals': 2,
    'Small businesses': 3,
    'Mid-sized businesses (charities and public bodies)': 4,
    'Large businesses': 5
}
table_b['segment_order'] = table_b['segment'].map(segment_order)
table_b = table_b.sort_values('segment_order').reset_index(drop=True)

print("Table B:")
display(table_b)
```

Table C — Sector Trend Table (time series, 2018-19 to 2023-24)

This answers: How has sector-level CT liability and company population shifted over 6 years?

Building the long-format trend table

```python
# Builds Table C: a sector-level trend table over time
# Converts the section liability data from wide to long format
liability_long = t8_section.melt(
    id_vars='sector_code',
    value_vars=year_cols,
    var_name='year_col',
    value_name='ct_liability_gbp_m'
 )
liability_long['year'] = liability_long['year_col'].str.replace('liability_', '').str.replace('_', '-')
liability_long = liability_long.drop(columns='year_col')

# Converts the company count data from wide to long format
companies_long = t9_section.melt(
    id_vars='sector_code',
    value_vars=company_year_cols,
    var_name='year_col',
    value_name='companies_liable'
 )
companies_long['year'] = companies_long['year_col'].str.replace('companies_', '').str.replace('_', '-')
companies_long = companies_long.drop(columns='year_col')

# Merging the liability and company count tables by sector and year
table_c = liability_long.merge(companies_long, on=['sector_code', 'year'], how='left')

# Adding sector names for readability
sector_names = table_a[['sector_code', 'sector_name']].drop_duplicates()
table_c = table_c.merge(sector_names, on='sector_code', how='left')

# Calculating liability per company (in thousands of pounds)
table_c['liability_per_company_gbp_k'] = (
    (table_c['ct_liability_gbp_m'] * 1000) / table_c['companies_liable']
).round(1)

# Excluding the special 'U' category and sorting the rows
table_c = table_c[table_c['sector_code'] != 'U']
table_c = table_c.sort_values(['sector_code', 'year']).reset_index(drop=True)

print("Table C shape:", table_c.shape)
display(table_c.head(20))
```

Building the Master Dashboard Dataframe

Joining everything into one wide, dashboard-ready table at the sector × year level, with the 2024-25 customer segment KPIs as a separate reference table

```python
# Builds the master sector dashboard table from the long-form trend table
# Uses table_c which already contains sector-by-year liability and company counts
master_sector = table_c.copy()

# Adds the latest 2023-24 snapshot ratios from table_a
snapshot_cols = [
    'sector_code', 'effective_ct_rate_pct', 'loss_ratio_pct',
    'ct_per_company_gbp_k', 'ct_share_pct', 'ct_rank'
 ]
snapshot = table_a[snapshot_cols].copy()
snapshot['year'] = '2023-24'  # attach the latest year label
master_sector = master_sector.merge(snapshot, on=['sector_code', 'year'], how='left')

# Reordering columns to make the output easier to read
col_order = [
    'year', 'sector_code', 'sector_name',
    'ct_liability_gbp_m', 'companies_liable', 'liability_per_company_gbp_k',
    'effective_ct_rate_pct', 'loss_ratio_pct',
    'ct_share_pct', 'ct_rank'
 ]
master_sector = master_sector[col_order]

print("Master sector dataframe shape:", master_sector.shape)
display(master_sector.head(20))
```
